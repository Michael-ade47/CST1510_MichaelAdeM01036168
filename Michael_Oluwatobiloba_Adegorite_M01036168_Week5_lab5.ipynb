### Activity 3: Temperature Logger

Write a program that:
1. Asks the user for temperatures for 5 days
2. Writes each temperature to "temps.txt" with the day number
   Format: "Day 1: 22°C"
3. Reads the file and calculates the average temperature

<hr>

`````python
days=5
temperatures=[]
with open("CST1510/temps.txt", mode="w") as f8:
    for i in range (days):
        temperature=int(input(f"What is the temperature for Day:{i+1}"))
        f8.write(f"Day{i+1},{temperature}\n")
        temperatures.append(temperature)
        total=sum(temperatures)
        Average=total/5
        
with open("CST1510/temps.txt", mode="r") as f9:
  temps=f9.readlines()
  for line in temps:
      day, temp = line.strip().split(",")
      print(f"{day}: {temp}°C")
print("The average temeperature for the five days is",Average,"°C")
        
`````

### **Activity 5: Unique Hobbies Set**

- Hobbies only Alex enjoys
- All unique hobbies combined

---

Create a dictionary for 3 students and their scores:

- Update Bob's score to 95
- Print all students with scores above 80

---

### Activity 17: Counting Words in Les_Brown.txt

Read the content from `Les_Brown.txt`, and then count the number of occurrences of each word in the text. Write the word count results to a new file called `word_count.txt`.

**Steps:**
1. Open and read the contents of `Les_Brown.txt`.
2. Split the content into words and count the frequency of each word.
3. Write the word count in the format: `word: count` to a new file called `word_count.txt`.

<hr>

`````python
with open("CST1510/Les_Brown.txt", mode="r") as f:#open file in specified file path with mode as  read
   data=f.read()#read the contents in Les_Brown.txt but by using its short name f and store it the variable data so we can print it
word_list = data.split()#splits the text into a list of words
word_frequency={}#word_frequency is an empty dictionary that will store each words frequency
for current_word in word_list:
    current_word=current_word.lower()#for counting make all the words lower case
    word_frequency[current_word]=word_frequency.get(current_word,0)+1#everytime current_word is found add 1 to  word frequncy
with open("CST1510/word_count.txt", mode="w") as f2:#open word_count text file if made else create it in the specified path and write in it
    for current_word, count in word_frequency.items():# Loop through each word and its count and write them to f2
        f2.write(f"{current_word}: {count}\n")

print("Word count written into word_count.txt file")#message to inform user thatthe count has been written into the word_count.txt file
with open("CST1510/word_count.txt", mode="r") as f3:#open file in specified file path with mode as read
   test=f3.read()#read the contents in word_count.txt but by using its short name f3 and store it the variable test so we can print it
print(test)
    
`````
