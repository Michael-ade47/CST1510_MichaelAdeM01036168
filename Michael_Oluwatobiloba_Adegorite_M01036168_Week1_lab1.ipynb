### Excellent 3. IT: Path Decomposition Report

Ask for a file or directory path and print its absolute path, parent, name, stem, and suffix.

**Example Output:**
```

Parent: /home/user/project
Name: service.log
Stem: service

```


*Optional hint:* `from pathlib import Path`


````python
# Your code here
#python to Ask for a file or directory path and print its absolute path, parent, name, stem, and suffix.


from pathlib import Path #imports path library

user_input=input('pls enter a file name or directory path:')# collects  user input  for file name

file_path=Path(user_input)#collects user input for file path

print('absolute path:',file_path.resolve())#reteturns the absolute path

print('Parent:',file_path.resolve().parent)#retuns the parent 

print('Name:',file_path.name)#retuns the name

print('stem:',file_path.stem)#returns the stem

print('suffix:',file_path.suffix)#returns the suffix
          
````

### Stretch 4. Payroll Statement

Ask for: employee name, hours worked, hourly rate, federal tax rate, state tax rate.
Print a formatted payroll slip with gross pay, each deduction, and net pay (2 d.p.).

**Example Output:**
```
Employee: Smith
Gross pay: £97.50
Federal withholding: £19.50
State withholding: £8.78
Net pay: £69.22
```


````python
# Your code here
Employee_name=input('Enter your name')

hours_worked=int(input('How many hours do you work'))

hourly_rate=int(input('How much do you earn  per hour'))

federal_tax=float(input('How much is thhe federal tax rate?'))

state_tax=float(input('How much is thhe state tax rate?'))

federal_tax_rate=federal_tax/100

state_tax_rate=state_tax/100

Gross_pay=hours_worked*hourly_rate

Federal_withholding=federal_tax_rate*Gross_pay

state_withholding=state_tax_rate*Gross_pay

Net_pay=Gross_pay-(Federal_withholding+State_withholding)
print('\nEmployee:',Employee_name)
print(f'Gross pay: £{Gross_pay:.2f}')
print(f'Federal withholding: £{Federal_withholding:.2f}')
print(f'State withholding: £{state_withholding:.2f}')
print(f'Net pay: £{Net_pay:.2f}')
````
