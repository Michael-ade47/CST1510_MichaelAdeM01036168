9. (Optional)
Basic Regex for Pattern Matching
Import the re module and use it to search for an email address in a string.
Use regex to find all the words that start with a capital letter in a sentence.

````python
import re #imports re library
text=" My name is Michael  e-mail:Michaeladegorite14@gmail.com"#assigning string to the variable text
pattern = r" e-mail:(\w+@\w+\.\w{3})" #defining the pattern
match=re.search(pattern,text)#using the search for  pattern within text
if match:
    print("Found a match:",match.group())#prints the  matched items
else:#if match is false prints No match
    print("No match")
capital_pattern=r"[A-Z][a-zA-Z]*"#pattern is Capital letter then small or capital letters
capital_word=re.findall(capital_pattern,text)#finds all items that have the patttern
print("The words that begin with capitals are:",capital_word)#prints the words that begin with capital

    
````

15. (Binary to decimal) Write a function that parses a binary number as a string and convert it into a decimal integer. Use the function header: def binaryToDecimal(binaryString): For example, binary string 10001 is 17 (1 24 + 0
23 + 0 22 + 0 21 + 1 * 20 = 17)

So, binaryToDecimal("10001") returns 17.

Write a test program that prompts the user to enter a binary string and displays the corresponding decimal integer value.

````python
binaryString=input("Enter a number in binary to be converted back to decimal")
def binaryToDecimal(binaryString):#defining the function binaryToDecimal
    decimal=0 
    length_binary=len(binaryString)#using len to get the length
    print(length_binary)#printing the length of binaryString
    for i, bit_char in enumerate(binaryString[::-1]):#using reverse slicing to count from the rigth
        bit=int(bit_char)#making bit_char an integer
        decimal+= bit *(2**i)#decimal=decimal plus bit multiplied by two raised to the power of i
        # decimal increase by bit(1 or 0) multiplied by two raised to the power of i
    print(binaryString,"in decimal is",decimal)#print output
        
binaryToDecimal(binaryString)
````
