````python


password=input("Enter password") #collects the user input and store it in the variable password
actual_password='MDX123' #assigning a value  to the variable  actual_password
if password==actual_password: #checks is password is equal to actual_password
    print('password was guessed 100% accuracy') #prints this when password is the same as the actual_password
elif password.lower()==actual_password.lower(): #changes the user input to lowercase from higher case and then checks is the lowercase 
    #password is the same as lowercase actual_password
            print("password was not guessed  50% accuracy")#prints this when the lowercase password is the same as lowercase actual_password
else: #else statement
    matches=0 #assigning the value zero to the variable matches
    length=len(actual_password) # uses the function len to find the amount of characters in  actual_password  and store it in the length
    for i in  range (min(len(password),length)):
        if password[i]==actual_password[i]:#checks if i in  password is equal to the i in  actual_password
            matches=matches+1 #updates matches by  adding 1 to matches  when the loop runs
accuracy=(matches/length)*100 # assigning the value  of matches divided by length  multiplied by 100  to the variable accuracy
print(f"The password is {accuracy:.0f}% accurate")   #prints  the accuracy   with zero decimal places

````
